### Right off the bat I can see there are a couple columns that need correcting to the propert format/date-type. 

#### Going to get rid of birth just because already have age

```python
df.drop(['BIRTH'], axis=1, inplace=True) 
```

```python
df.info()
```

#### Convert monetary value columns to floats

```python
money_columns_convert_to_float = ["INCOME", "HOME_VAL", "BLUEBOOK", 
                                 "OLDCLAIM", "CLM_AMT"]
```

```python
for col in money_columns_convert_to_float:
    df[col] = df[col].str.replace(",", "").str.replace("$", "")
    df[col] = df[col].astype(float)  
    
```

```python
df.info()
```

#### We have 2645 rows with a null value

```python
df[df.isnull().any(axis=1)].shape
```

#### Null values in YOJ (float64), Income (float64), Home_Val (home_val), OCCUPATION (object/categorical), CAR_AGE (float64)

```python
df.info()
```

##### Convert 

```python
numeric_columns_with_null = [
    "YOJ", "INCOME", "HOME_VAL","CAR_AGE", "AGE"
]
```

```python
categorical_columns_with_null = [
    "OCCUPATION"
]
```

```python
for col in numeric_columns_with_null:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df[df['OCCUPATION'].isnull()]
```

```python
df['OCCUPATION'].value_counts(normalize=True)
```

##### Fill with the mode of the occupation column

```python
for col in categorical_columns_with_null:
    df[col].fillna(df[col].value_counts().index[0], inplace=True)  
```

```python
df['OCCUPATION'].value_counts(normalize=True)
```

```python
df.info()
```

```python
df.head()
```

```python
convert_column_to_binary = ["PARENT1", "MSTATUS", "RED_CAR", "REVOKED"]
```

```python
for col in convert_column_to_binary:
    print(col, df[col].value_counts())
```

```python
for col in convert_column_to_binary:
    df[col] = df[col].str.replace("z_No", "0")
    df[col] = df[col].str.replace("No", "0")
    
    df[col] = df[col].str.replace("no", "0")
    
    df[col] = df[col].str.replace("Yes", "1")
    df[col] = df[col].str.replace("yes", "1")
    df[col] = df[col].astype(int)
```

## Classification of claim or no claim first

```python
y,X = df['CLAIM_FLAG'].copy(), df.drop(["CLAIM_FLAG", "CLM_AMT", "CLM_FREQ", 
               "REVOKED"],axis=1).copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.model_selection import cross_val_score, RandomizedSearchCV
from lightgbm import LGBMClassifier
from sklearn.metrics import (accuracy_score, recall_score, auc, 
                            confusion_matrix, average_precision_score, roc_auc_score)
```

## Try a baseline light gradient boosting machine classifier, pretty solid results with model performing much better than random guessing (roc_auc = .50)

```python
base_lgbm = LGBMClassifier()
base_lgbm.fit(X_train, y_train)
preds = base_lgbm.predict_proba(X_test)[:,1]
print(f"Average Precision Score: {average_precision_score(y_test, preds)}")
print(f"ROC_AUC Score: {roc_auc_score(y_test, preds)}")
```

```python
params = {
    'application': 'binary', # for binary classification
#     'num_class' : 1, # used for multi-classes
    'boosting': 'gbdt', # traditional gradient boosting decision tree
    'num_iterations': 100, 
    'learning_rate': 0.05,
    'num_leaves': 62,
    'device': 'cpu', # you can use GPU to achieve faster learning
    'max_depth': -1, # <0 means no limit
    'max_bin': 510, # Small number of bins may reduce training accuracy but can deal with over-fitting
    'lambda_l1': 5, # L1 regularization
    'lambda_l2': 10, # L2 regularization
    'metric' : 'binary_error',
    'subsample_for_bin': 200, # number of samples for constructing bins
    'subsample': 1, # subsample ratio of the training instance
    'colsample_bytree': 0.8, # subsample ratio of columns when constructing the tree
    'min_split_gain': 0.5, # minimum loss reduction required to make further partition on a leaf node of the tree
    'min_child_weight': 1, # minimum sum of instance weight (hessian) needed in a leaf
    'min_child_samples': 5# minimum number of data needed in a leaf
}
```

```python
lgbm_model = LGBMClassifier(boosting_type= 'gbdt', 
          objective = 'binary', 
          n_jobs = 5, 
          silent = True,
          max_depth = params['max_depth'],
          max_bin = params['max_bin'], 
          subsample_for_bin = params['subsample_for_bin'],
          subsample = params['subsample'], 
          min_split_gain = params['min_split_gain'], 
          min_child_weight = params['min_child_weight'], 
          min_child_samples = params['min_child_samples'])
```

## Parameter grid to find best hyperparameters

```python
rndmParams = {
    'learning_rate': [0.005, 0.01],
    'n_estimators': [8,16,24],
    'num_leaves': [6,8,12,16], # large num_leaves helps improve accuracy but might lead to over-fitting
    'boosting_type' : ['gbdt', 'dart'], # for better accuracy -> try dart
    'objective' : ['binary'],
    'max_bin':[255, 510], # large max_bin helps improve accuracy but might slow down training progress
    'random_state' : [500],
    'colsample_bytree' : [0.64, 0.65, 0.66],
    'subsample' : [0.7,0.75],
    'reg_alpha' : [1,1.2],
    'reg_lambda' : [1,1.2,1.4],
    }
```

## Random search faster to train than grid search

```python
randomCV = RandomizedSearchCV(lgbm_model, rndmParams, verbose=2, n_jobs=-1, cv=4, scoring="average_precision") 
```

```python
randomCV.fit(X_train, y_train)
```

## Depending on the threshold we set, on average when the model predicts the positive class, it does so ~55 % correctly on average which is solid considering the imbalanced classes with close to 3:1 ratio of negative class to positive class. In real world scenario, considering these customers were already approved by underwriters, I would set a high positive class threshold in order to increase precision with the reasoning that I do want to identify customers likely to incur a claim; however, I do not want to turn away a significant portion of business that underwriters already deemed a reasonable risk to take on. I only want to turn away the riskiest of these businesses that were accepted


```python
print(randomCV.best_params_)
print(f"Mean Average Precision {round(randomCV.best_score_,3)}")  
```

```python
lgbm_model_best = LGBMClassifier(**randomCV.best_params_) 
```
